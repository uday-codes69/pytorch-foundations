# Pytorch tensors & numpy

Numpy is a popular scientific python numerical computing library

Data in Numpy, want in Pytorch tensor -> torch.from_numpy(ndarray)

Pytorch Tensor -> Numpy -> torch.Tensor.numpy()

```python
import torch
import numpy as np

array = np.arange(1.0,8.0)
tensor = torch.from_numpy(array) #warning: when converting from numpy -> pytorch, pytorch numpy's default datatype of float64 unless specified otherwise
array,tensor
```

```python
#change the value of array, what will this do to 'tensor'?
array = array + 1
array, tensor
```

```python
array.dtype
```

```python
tensor = torch.arange(1.0, 8.0)
tensor.dtype
```

```python
# Tensor to Numpy array
tensor = torch.ones(7)
nump_tensor = tensor.numpy()
tensor, nump_tensor
```

```python
# Change the tensor, what happen to numpy_tensor?
tensor = tensor+1
tensor, nump_tensor
```

## Reproducbility (trying to take random out of random)

In short how a neural network learns

start with number -> take operations -> update random numbers to try and make them better representation of the data -> again -> again and again

To reduce the randomness in neural networks and pytorch come the concept of a **random seed**

```python
import torch

#create two random tensors
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

*torch.manual_seed() fixes the starting point of PyTorch’s random number generator, ensuring reproducible random tensors across runs.*

**Same seed → same “random” sequence. &&
Different seed → different sequence**

```python
 1 == 1
```

```python
import torch

RANDOM_SEED = 47
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3,4)


RANDOM_SEED = 47
torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3,4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
